```python
import numpy as np
arr = np.array([[1, 2], [3, 4]])
print("2D Array:\n", arr)
print("Element [1,1]:", arr[1,1])
```

```python
import numpy as np
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print("Matrix Addition:\n", a + b)
```

```python
import numpy as np
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print("Matrix Multiplication:\n", np.dot(a, b))
```

```python
import numpy as np
a = np.array([[1,2],[3,4]])
print("Determinant:", np.linalg.det(a))
```

```python
import numpy as np
a = np.array([[1,2],[3,4]])
print("Inverse:\n", np.linalg.inv(a))
```

```python
import numpy as np
arr = np.array([1, 2, 3, 4])
print("Original Array:", arr)
print("Transpose:", arr.T)
print("Shape:", arr.shape)
```

```python
import numpy as np
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print("Original Matrix:\n", arr)
print("Transpose:\n", arr.T)
print("Original Shape:", arr.shape)
print("Transpose Shape:", arr.T.shape)
```

```python
import numpy as np
arr1 = np.array([3, 4])
print("1D Array:", arr1)
print("Norm of 1D:", np.linalg.norm(arr1))
```

```python
import numpy as np
arr2 = np.array([[1, 2],
                 [3, 4]])
print("\n2D Array:\n", arr2)
print("Norm of 2D:", np.linalg.norm(arr2))
```
